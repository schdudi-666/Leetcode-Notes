```python
from collections import defaultdict, deque

class Solution:
    def ladderLength(self, beginWord: str, endWord: str, wordList: list[str]) -> int:
        # è‹¥ç›®æ ‡å•è¯ä¸åœ¨è¯è¡¨ä¸­ï¼Œç›´æ¥è¿”å› 0
        if endWord not in wordList:
            return 0

        # 1ï¸âƒ£ æ„å»º pattern å­—å…¸
        d = defaultdict(list)
        for word in wordList:
            for i in range(len(word)):
                d[word[:i] + '*' + word[i+1:]].append(word)

        # 2ï¸âƒ£ åˆå§‹åŒ– BFS é˜Ÿåˆ—å’Œè®¿é—®é›†åˆ
        q = deque([(beginWord, 1)])  # (å½“å‰å•è¯, å½“å‰æ­¥æ•°)
        visited = {beginWord}

        # 3ï¸âƒ£ BFS ä¸»å¾ªç¯
        while q:
            word, step = q.popleft()
            for i in range(len(word)):
                pattern = word[:i] + '*' + word[i+1:]
                for nxt in d[pattern]:
                    if nxt == endWord:
                        print(f"æ‰¾åˆ°è·¯å¾„ï¼Œé•¿åº¦ä¸º {step + 1}")
                        return step + 1
                    if nxt not in visited:
                        visited.add(nxt)
                        q.append((nxt, step + 1))
                d[pattern] = []  # æ¸…ç©ºä»¥åŠ é€Ÿ
        print("æœªæ‰¾åˆ°è·¯å¾„")
        return 0


# ğŸ” æµ‹è¯•ç¤ºä¾‹
solution = Solution()
beginWord = "hit"
endWord = "cog"
wordList = ["hot", "dot", "dog", "lot", "log", "cog"]

ans = solution.ladderLength(beginWord, endWord, wordList)
print(f"\næœ€ç»ˆç»“æœï¼šæœ€çŸ­è½¬æ¢åºåˆ—é•¿åº¦ = {ans}")
```

æœ¬é¢˜çš„æ€è·¯ä¸ä¸Šé¢˜ç±»ä¼¼ ä½†æ˜¯æˆ‘ä»¬ä½¿ç”¨æšä¸¾æ³• å¯¹æ¯ä¸ªåŸºå› ä½è¿›è¡Œå››ç§ä¸åŒçš„çªå˜ å¦‚æœçªå˜åŸºå› åœ¨bankä¸­ä¸”ä¸åœ¨visitedé‡Œåˆ™ä¸‹ä¸€æ¬¡å¾ªç¯

```python
from collections import deque

class Solution:
    def minMutation(self, startGene: str, endGene: str, bank: list[str]) -> int:
        # å°†åŸºå› åº“è½¬æ¢ä¸ºé›†åˆï¼Œæé«˜æŸ¥æ‰¾æ•ˆç‡
        bank = set(bank)
        if endGene not in bank:
            return -1

        # BFS é˜Ÿåˆ—ï¼šå­˜æ”¾ (å½“å‰åŸºå› , å½“å‰æ­¥æ•°)
        q = deque([(startGene, 0)])
        visited = {startGene}  # è®°å½•å·²è®¿é—®çŠ¶æ€

        while q:
            curr, step = q.popleft()
            if curr == endGene:
                print(f"æ‰¾åˆ°ç›®æ ‡åŸºå› : {curr}, æœ€å°‘çªå˜æ¬¡æ•° = {step}")
                return step

            # éå†æ¯ä¸€ä½ï¼Œå°è¯•4ç§ç¢±åŸºæ›¿æ¢
            for i in range(len(curr)):
                for c in "ACGT":
                    if c != curr[i]:
                        new_gene = curr[:i] + c + curr[i+1:]
                        if new_gene in bank and new_gene not in visited:
                            visited.add(new_gene)
                            q.append((new_gene, step + 1))
            
            # æ‰“å°å½“å‰å±‚æœç´¢çŠ¶æ€ï¼ˆå¯é€‰ï¼‰
            print(f"ç¬¬ {step} æ­¥æœç´¢å®Œæˆï¼Œå½“å‰åŸºå› : {curr}, é˜Ÿåˆ—é•¿åº¦: {len(q)}")

        print("æ— æ³•çªå˜åˆ°ç›®æ ‡åŸºå› ã€‚")
        return -1


# ğŸ§ª æµ‹è¯•æ ·ä¾‹
solver = Solution()
start = "AACCGGTT"
end = "AAACGGTA"
bank = ["AACCGGTA", "AACCGCTA", "AAACGGTA"]

solver.minMutation(start, end, bank)
```
