# 打家劫舍2

### 思路

本题的思路是是用动态规划 在这里我们展示两种写法 一种是展开动态规划 另一种是用子函数解决

### 代码

```python
# 方法一：显式双 DP 表
class Solution1:
    def rob(self, nums):
        if len(nums) <= 2:
            return max(nums)
        dpf = [0] * len(nums)
        dpl = [0] * len(nums)
        dpf[0], dpf[1] = nums[0], max(nums[0], nums[1])
        dpl[0], dpl[1] = nums[1], max(nums[1], nums[2])
        for i in range(2, len(nums)-1):
            dpf[i] = max(dpf[i-1], dpf[i-2] + nums[i])
        for i in range(2, len(nums)-1):
            dpl[i] = max(dpl[i-1], dpl[i-2] + nums[i+1])
        return max(max(dpf), max(dpl))

# 方法二：子函数封装法
class Solution2:
    def rob(self, nums):
        if len(nums) <= 2:
            return max(nums)
        def rob_line(houses):
            dp = [0] * len(houses)
            dp[0] = houses[0]
            dp[1] = max(houses[0], houses[1])
            for i in range(2, len(houses)):
                dp[i] = max(dp[i-1], dp[i-2] + houses[i])
            return dp[-1]
        return max(rob_line(nums[:-1]), rob_line(nums[1:]))

# 测试样例
nums_list = [
    [2,3,2],
    [1,2,3,1],
    [1,2,3],
    [1,7,9,2],
    [200,3,140,20,10]
]

print("213. House Robber II 比较输出：\n")
for nums in nums_list:
    ans1 = Solution1().rob(nums)
    ans2 = Solution2().rob(nums)
    print(f"输入: {nums}")
    print(f"方法一输出: {ans1}")
    print(f"方法二输出: {ans2}")
    print("-"*40)
```

### 类似题目（91. 解码方法）

### 思路

本题的思路是用动态规划的方法在递推时考虑两种不同情况：本字符是否可以单独编码？以及本字符和上个字符可否共同编码？

### 代码

```python
class Solution:
    def numDecodings(self, s: str) -> int:
        if s[0] == '0':
            return 0
        dp = [0] * (len(s) + 1)
        dp[0] = 1
        dp[1] = 1
        for i in range(2, len(s)+1):
            # 单字符解码（如果当前位不是 '0'）
            if s[i-1] != '0':
                dp[i] += dp[i-1]
            # 双字符解码（如果前两位合法）
            if 10 <= int(s[i-2:i]) <= 26:
                dp[i] += dp[i-2]
        return dp[len(s)]

tests = [
    "12",    # AB, L → 2
    "226",   # BZ, VF, BBF → 3
    "06",    # 无法解码 → 0
    "11106", # AAJF, KJF → 2
    "27"     # BG → 1
]

print("91. Decode Ways 结果输出：\n")
for t in tests:
    print(f"输入: {t} → 输出: {Solution().numDecodings(t)}")
```
