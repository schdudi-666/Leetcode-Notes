# 数组中第K大的元素

### 思路

这题本质是取第 k 大元素，可以通过排序后取 nums[-k] 得到。

使用快速排序（QuickSort）对数组进行排序：选 pivot，将数组划分为左右两边。

左边 ≤ pivot，右边 ≥ pivot，然后递归分别排序。

最终整个数组有序，直接 return 排序后数组中的倒数第 k 个即可。

QuickSort 平均 O(n log n)，代码简单直观但不是此题最优解（QuickSelect 才是 O(n)）。

### 代码

```python
# QuickSort version for LeetCode 215 — Jupyter Notebook friendly

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    mid  = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    
    return quicksort(left) + mid + quicksort(right)


def findKthLargest(nums, k):
    sorted_nums = quicksort(nums)
    return sorted_nums[-k]


# -----------------------------
# Example test
# -----------------------------
nums = [3,2,1,5,6,4]
k = 2

result = findKthLargest(nums, k)
print("第 {} 大的数是：{}".format(k, result))
```

### 类似问题（658. 找到K个最近元素）

### 思路

因为 arr 已经排好序，所以先用二分（或双指针）找到 最靠近 x 的位置。

设置两个指针：left 在该位置左边，right 在该位置右边。

每次比较 abs(arr[left] - x) 和 abs(arr[right] - x) 哪个更小。

小的那一侧向外扩展一步（left-- 或 right++）。

扩展 k 次后，left+1 到 right-1 即是答案；排序后返回。

### 代码

```python
def findClosestElements(arr, k, x):
    # 1. 二分查找最近的位置
    import bisect
    right = bisect.bisect_left(arr, x)
    left = right - 1

    # 2. 扩散 k 次
    while k > 0:
        if left < 0:
            right += 1
        elif right >= len(arr):
            left -= 1
        elif abs(arr[left] - x) <= abs(arr[right] - x):
            left -= 1
        else:
            right += 1
        k -= 1

    # 3. 返回窗口内的元素
    return arr[left+1:right]


# ---------------------
# Example test
# ---------------------
arr = [1, 2, 3, 4, 5]
x = 3
k = 4
print(findClosestElements(arr, k, x))
```
