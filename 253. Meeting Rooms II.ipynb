```python
import heapq

class Solution:
    def minMeetingRooms(self, intervals):
        # æŒ‰ä¼šè®®å¼€å§‹æ—¶é—´æ’åº
        intervals.sort(key=lambda x: x[0])

        min_heap = []  # å­˜æ”¾æ­£åœ¨è¿›è¡Œçš„ä¼šè®®çš„ç»“æŸæ—¶é—´

        for start, end in intervals:
            # å¦‚æœæœ€æ—©ç»“æŸçš„ä¼šè®®å·²ç»“æŸï¼Œå¤ç”¨è¯¥ä¼šè®®å®¤
            if min_heap and min_heap[0] <= start:
                heapq.heappop(min_heap)
            # åŠ å…¥å½“å‰ä¼šè®®çš„ç»“æŸæ—¶é—´
            heapq.heappush(min_heap, end)

        print(f"ä¼šè®®å®‰æ’: {intervals}")
        print(f"æ‰€éœ€ä¼šè®®å®¤æ•°é‡: {len(min_heap)}")

# ç¤ºä¾‹æµ‹è¯•
intervals = [[0, 30], [5, 10], [15, 20]]
sol = Solution()
sol.minMeetingRooms(intervals)
```

```python
import heapq
from collections import Counter, deque

class Solution:
    def leastInterval(self, tasks, n):
        # 1ï¸âƒ£ ç»Ÿè®¡ä»»åŠ¡å‡ºç°æ¬¡æ•°
        d = Counter(tasks)
        
        # 2ï¸âƒ£ å»ºç«‹æœ€å¤§å †ï¼ˆæ¬¡æ•°å–è´Ÿæ•°ï¼‰
        max_heap = []
        for task, count in d.items():
            heapq.heappush(max_heap, (-count, task))
        
        # 3ï¸âƒ£ å†·å´é˜Ÿåˆ—ï¼šå­˜ (cnt, task, ready_time)
        queue = deque()
        time = 0
        
        # 4ï¸âƒ£ ä¸»å¾ªç¯
        while max_heap or queue:
            time += 1

            # å¦‚æœå †éç©ºï¼Œå–å‡ºä¸€ä¸ªä»»åŠ¡æ‰§è¡Œä¸€æ¬¡
            if max_heap:
                cnt, task = heapq.heappop(max_heap)
                cnt += 1  # å› ä¸ºæ˜¯è´Ÿæ•°ï¼Œæ‰€ä»¥ +1 è¡¨ç¤ºæ‰§è¡Œäº†ä¸€æ¬¡
                # å¦‚æœä»»åŠ¡è¿˜æœ‰å‰©ä½™æ¬¡æ•°ï¼Œæ”¾å…¥å†·å´é˜Ÿåˆ—
                if cnt != 0:
                    queue.append((cnt, task, time + n))
            
            # å¦‚æœå†·å´é˜Ÿåˆ—çš„ä»»åŠ¡å·²å‡†å¤‡å¥½ï¼Œæ”¾å›å †
            if queue and queue[0][2] == time:
                cnt, task, ready_time = queue.popleft()
                heapq.heappush(max_heap, (cnt, task))
        
        print(f"ä»»åŠ¡: {tasks}, å†·å´æ—¶é—´ n={n}")
        print(f"å®Œæˆæ‰€æœ‰ä»»åŠ¡çš„æœ€çŸ­æ—¶é—´: {time}")
        return time


# ğŸ” ç¤ºä¾‹æµ‹è¯•
tasks = ["A", "A", "A", "B", "B", "B"]
n = 2
sol = Solution()
sol.leastInterval(tasks, n)
```
