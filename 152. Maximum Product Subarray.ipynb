```python
class Solution:
    def maxProduct(self, nums):
        n = len(nums)
        maxdp = [0] * n
        mindp = [0] * n
        maxdp[0] = nums[0]
        mindp[0] = nums[0]

        for i in range(1, n):
            maxdp[i] = max(nums[i], maxdp[i-1] * nums[i], mindp[i-1] * nums[i])
            mindp[i] = min(nums[i], maxdp[i-1] * nums[i], mindp[i-1] * nums[i])

        return max(maxdp)

nums = [2, 3, -2, 4, -1]
res = Solution().maxProduct(nums)
print("è¾“å…¥æ•°ç»„ï¼š", nums)
print("æœ€å¤§ä¹˜ç§¯ï¼š", res)

n = len(nums)
maxdp = [0] * n
mindp = [0] * n
maxdp[0] = nums[0]
mindp[0] = nums[0]

for i in range(1, n):
    maxdp[i] = max(nums[i], maxdp[i-1] * nums[i], mindp[i-1] * nums[i])
    mindp[i] = min(nums[i], maxdp[i-1] * nums[i], mindp[i-1] * nums[i])

print("\nðŸ“ˆ çŠ¶æ€å˜åŒ–ï¼š")
print("Idx | Num | MaxDP | MinDP")
print("-" * 30)
for i in range(n):
    print(f"{i:>3} | {nums[i]:>4} | {maxdp[i]:>6} | {mindp[i]:>6}")
```

```python
class Solution:
    def getMaxLen(self, nums):
        pos = [0] * len(nums)
        neg = [0] * len(nums)
        pos[0] = 1 if nums[0] > 0 else 0
        neg[0] = 1 if nums[0] < 0 else 0

        for i in range(1, len(nums)):
            num = nums[i]
            if num > 0:
                pos[i] = pos[i-1] + 1
                neg[i] = neg[i-1] + 1 if neg[i-1] > 0 else 0
            elif num < 0:
                pos[i] = neg[i-1] + 1 if neg[i-1] > 0 else 0
                neg[i] = pos[i-1] + 1
            else:
                pos[i] = 0
                neg[i] = 0

        return max(pos)

nums = [1, -2, -3, 4]
res = Solution().getMaxLen(nums)
print("è¾“å…¥æ•°ç»„ï¼š", nums)
print("æœ€é•¿æ­£ç§¯å­æ•°ç»„é•¿åº¦ï¼š", res)

pos = [0] * len(nums)
neg = [0] * len(nums)
pos[0] = 1 if nums[0] > 0 else 0
neg[0] = 1 if nums[0] < 0 else 0

for i in range(1, len(nums)):
    num = nums[i]
    if num > 0:
        pos[i] = pos[i-1] + 1
        neg[i] = neg[i-1] + 1 if neg[i-1] > 0 else 0
    elif num < 0:
        pos[i] = neg[i-1] + 1 if neg[i-1] > 0 else 0
        neg[i] = pos[i-1] + 1
    else:
        pos[i] = 0
        neg[i] = 0

print("\nðŸ“ˆ çŠ¶æ€å˜åŒ–ï¼š")
print("Idx | Num | PosLen | NegLen")
print("-"*33)
for i in range(len(nums)):
    print(f"{i:>3} | {nums[i]:>4} | {pos[i]:>7} | {neg[i]:>7}")
```
