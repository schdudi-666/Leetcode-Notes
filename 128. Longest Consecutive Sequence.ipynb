# 最长连续序列

### 思路

1. 将数组转换为集合 `setnums`，以便 O(1) 判断某个数字是否存在。
2. 遍历集合中的每个元素 `num`，只有当 `num-1` 不在集合中时，将其视为某个连续序列的起点。
3. 从起点向右不断检查 `num+1, num+2, ...` 是否存在于集合中，使用 while 循环统计该序列长度。
4. 在所有序列中维护一个最大长度 `lc`。
5. 返回最长连续序列的长度 `lc`，整体时间复杂度为 O(n)。

### 代码

```python
class Solution:
    def longestConsecutive(self, nums):
        lc = 0
        setnums = set(nums)

        for num in setnums:
            # 如果 num-1 不在集合中，它才是序列起点
            if num - 1 not in setnums:
                cc = 1
                curr = num
                # 向右扩展连续序列
                while curr + 1 in setnums:
                    curr += 1
                    cc += 1
                lc = max(lc, cc)

        return lc


# Jupyter 测试输出
solver = Solution()
nums = [100, 4, 200, 1, 3, 2]  # 最长连续序列是 [1,2,3,4]
solver.longestConsecutive(nums)
```

### 类似题目（1200. 最小绝对差）

### 思路

1. 将数组升序排序，最小绝对差一定出现在相邻元素之间（数学性质）。
2. 第一次遍历相邻元素，找到所有相邻差值中的最小值 `min_diff`。
3. 第二次遍历相邻元素，收集所有差值等于 `min_diff` 的数对 `[arr[i], arr[i+1]]`。
4. 因为数组已排序，所有输出对自动按字典序排好。
5. 排序 O(n log n)，扫描两次 O(n)，整体为最优复杂度。

### 代码

```python
class Solution:
    def minimumAbsDifference(self, arr):
        arr.sort()
        min_diff = float('inf')

        # 第一次遍历：找到最小差值
        for i in range(len(arr) - 1):
            min_diff = min(min_diff, arr[i+1] - arr[i])

        # 第二次遍历：收集所有符合最小差的 pair
        res = []
        for i in range(len(arr) - 1):
            if arr[i+1] - arr[i] == min_diff:
                res.append([arr[i], arr[i+1]])

        return res


# Jupyter 测试样例
solver = Solution()
arr = [4, 2, 1, 3]
solver.minimumAbsDifference(arr)
```
