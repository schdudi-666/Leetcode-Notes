# 换硬币

### 思路

目标：
求最少硬币数，使它们的总金额等于 amount。
若无法凑出，返回 -1。

状态定义：
dp[i] = 凑出金额 i 所需的最少硬币数。

初始化：

dp[0] = 0（金额 0 不需要硬币）

其余初始为 inf（表示暂时不可达）

状态转移：
对每个金额 i，尝试每个硬币 coin：

dp[i] = min(dp[i], dp[i - coin] + 1)   (if i >= coin)


含义：
如果我能凑出 i - coin，那再加上一枚 coin 就能凑出 i。

结果判断：
若 dp[amount] == inf → 无法凑出 → 返回 -1；
否则返回 dp[amount]。

### 代码

```python
# LeetCode 322. Coin Change

class Solution:
    def coinChange(self, coins: list[int], amount: int) -> int:
        if amount == 0:
            return 0

        dp = [float('inf')] * (amount + 1)
        dp[0] = 0

        for i in range(1, amount + 1):
            for coin in coins:
                if i - coin >= 0:
                    dp[i] = min(dp[i], dp[i - coin] + 1)

        return dp[-1] if dp[-1] != float('inf') else -1


# 示例测试
coins = [1, 2, 5]
amount = 11
sol = Solution()
result = sol.coinChange(coins, amount)

print("硬币种类:", coins)
print("目标金额:", amount)
print("最少硬币数:", result)
```

### 类似题目（279. 完全平方数）

### 思路

目标：
求最少的完全平方数，使它们的和等于 n。
例如 n = 13 → 9 + 4 → 最少 2 个。

状态定义：
dp[i] 表示凑出整数 i 所需的最少平方数个数。

初始化：

dp[0] = 0（0 不需要任何平方数）

其余为无穷大（float('inf')），表示暂不可达。

状态转移：
对每个整数 i，遍历所有小于等于 i 的平方数 p：

dp[i] = min(dp[i], dp[i - p] + 1)


意思是：如果我能凑出 i - p，
那再加上一个平方数 p 就能凑出 i。

答案：
dp[n] 即为最少平方数数量。

### 代码

```python
# LeetCode 279. Perfect Squares

class Solution:
    def numSquares(self, n: int) -> int:
        if n == 0:
            return 0
        
        dp = [float('inf')] * (n + 1)
        dp[0] = 0
        
        squares = [i * i for i in range(1, int(n ** 0.5) + 1)]
        
        for i in range(1, n + 1):
            for p in squares:
                if i - p >= 0:
                    dp[i] = min(dp[i], dp[i - p] + 1)
        
        return dp[-1]


# 示例测试
n = 13
sol = Solution()
result = sol.numSquares(n)

print("输入:", n)
print("最少平方数个数:", result)
```
