# 目标和

### 思路

把问题转成选出部分数字，使它们和为 P，其中 P = (S + target) / 2。

如果 (S + target) 是奇数或 P < 0，则答案为 0。

定义 dp[x]：表示用 nums 中的数凑出和 x 的方案数。

初始化 dp[0] = 1，表示“什么都不选”凑出 0 的方案。

对每个 num 从后往前更新 dp，最终答案是 dp[P]。

### 代码

```python
# 494. Target Sum  —— DP 背包版

nums = [1,1,1,1,1]
target = 3

# -----------------------
#      DP 核心逻辑
# -----------------------

S = sum(nums)
# 转化成功条件：S + target 必须是偶数，并且 P >= 0
if (S + target) % 2 != 0 or S < abs(target):
    print("答案 = 0")
else:
    P = (S + target) // 2   # 求子集和目标
    print("转换后的 P =", P)

    # dp[x] = 凑出和 x 的方案数
    dp = [0] * (P + 1)
    dp[0] = 1   # 和为 0 的方案：什么都不选

    for num in nums:
        # 背包必须逆序，因为每个数只能用一次
        for x in range(P, num - 1, -1):
            dp[x] += dp[x - num]

    print("dp 数组 =", dp)
    print("答案 = dp[P] =", dp[P])
```

### 类似题目（377. 合并和4）

### 思路

顺序不同算不同，是“排列型 DP”，属于完全背包问题。

定义 dp[i] 为凑出和 i 的排列数量。

dp[0] = 1（空序列）。

从 1 到 target 遍历 i，再遍历 nums：若 i ≥ num，则 dp[i] += dp[i - num]。

最终答案为 dp[target]。|

### 代码

```python
# LeetCode 377. Combination Sum IV

nums = [1, 2, 3]
target = 4

# -----------------------------
#       排列型完全背包 DP
# -----------------------------

dp = [0] * (target + 1)
dp[0] = 1   # 凑出 0 的方式：空序列

for i in range(1, target + 1):
    for num in nums:
        if i >= num:
            dp[i] += dp[i - num]

print("nums =", nums)
print("target =", target)
print("dp 数组 =", dp)
print("答案 dp[target] =", dp[target])
```
