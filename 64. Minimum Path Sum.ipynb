# 最小路径和

### 思路

用动态规划：dp[i][j] 表示从起点到 (i,j) 的最小路径和。
第一行和第一列先累加初始化；
之后每个格子的最优路径为 grid[i][j] + min(上方, 左方)；
最终结果是右下角的 dp[m-1][n-1]。

### 代码

```python
from typing import List

class Solution:
    def minPathSum(self, grid: List[List[int]]) -> int:
        m, n = len(grid), len(grid[0])
        dp = [[0 for _ in range(n)] for _ in range(m)]
        dp[0][0] = grid[0][0]

        # 初始化第一列
        for i in range(1, m):
            dp[i][0] = dp[i-1][0] + grid[i][0]

        # 初始化第一行
        for j in range(1, n):
            dp[0][j] = dp[0][j-1] + grid[0][j]

        # 状态转移
        for i in range(1, m):
            for j in range(1, n):
                dp[i][j] = grid[i][j] + min(dp[i-1][j], dp[i][j-1])

        return dp[m-1][n-1]

grid = [
    [1,3,1],
    [1,5,1],
    [4,2,1]
]
solution = Solution()
print("最小路径和为：", solution.minPathSum(grid))
```

### 类似题目（120. 三角形最小路径和）

### 思路

用动态规划自底向上：
让 dp[i][j] 表示从 (i,j) 到底部的最小路径和。
最底层初始化为 triangle 最后一行，
状态转移为 dp[i][j] = triangle[i][j] + min(dp[i+1][j], dp[i+1][j+1])，
最终结果为 dp[0][0]。

### 代码

```python
# LeetCode 120 - 三角形最小路径和 (Triangle Minimum Path Sum)

from typing import List

class Solution:
    def minimumTotal(self, triangle: List[List[int]]) -> int:
        m = len(triangle)
        # 构建与 triangle 形状相同的 DP 表
        dp = [[0] * len(row) for row in triangle]

        # 初始化：最底层的路径和就是它本身
        for j in range(len(triangle[m-1])):
            dp[m-1][j] = triangle[m-1][j]

        # 自底向上递推
        for i in range(m - 2, -1, -1):
            for j in range(len(triangle[i])):
                dp[i][j] = triangle[i][j] + min(dp[i + 1][j], dp[i + 1][j + 1])

        return dp[0][0]

triangle = [
     [2],
    [3,4],
   [6,5,7],
  [4,1,8,3]
]

solution = Solution()
result = solution.minimumTotal(triangle)
print("最小路径和为：", result)
```
