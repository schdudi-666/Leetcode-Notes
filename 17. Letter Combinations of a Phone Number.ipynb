# 电话号码字母组合

### 思路

1. 每位数字映射到若干字母（如 2→"abc"），组合要从每位中取一个字母拼起来。
2. 使用 BFS：每处理一位数字，就把当前所有前缀扩展成下一层前缀。
3. 队列初始为 [""]，表示还没有选择任何字母。
4. 对 digits 的每一位 digit：对队列中的每个前缀 prefix，尝试接上 digit 对应的所有字母。
5. 最后一层生成的所有前缀就是最终的全部字母组合。

### 代码

```python
from collections import deque

def letterCombinations(digits: str):
    if not digits:
        return []

    phone = {
        '2': 'abc',
        '3': 'def',
        '4': 'ghi',
        '5': 'jkl',
        '6': 'mno',
        '7': 'pqrs',
        '8': 'tuv',
        '9': 'wxyz'
    }

    q = deque([''])  # BFS 初始层

    for digit in digits:
        next_q = deque()
        while q:
            prefix = q.popleft()
            for ch in phone[digit]:
                next_q.append(prefix + ch)
        q = next_q  # 进入下一层

    return list(q)

# 测试
letterCombinations("25")
```

### 类似题目（78. 子集）

### 思路

1. 初始队列为 [[]]，代表目前只有一个空子集。
2. 按顺序遍历每个数字 num，每处理一个数字就扩展一层。
3. 对当前队列中的每个 subset，有两个选择：不选 num 或选 num。
4. 不选 num → subset 原样加入下一层；选 num → subset+[num] 加入下一层。
5. 全部数字处理完后，队列中的所有 subset 就是所有子集。


### 代码

```python
from collections import deque

def subsets(nums):
    # 初始只有一个空子集
    q = deque([[]])

    for num in nums:
        next_q = deque()

        # 扩展当前层的所有前缀
        while q:
            prefix = q.popleft()

            # 不选 num
            next_q.append(prefix)

            # 选 num
            next_q.append(prefix + [num])

        # 进入下一层
        q = next_q

    return list(q)


# 测试
subsets([1, 2, 3])
```
