# 最多水问题

## 重要：双指针必需保证数组有序

### 思路

本题的思路是使用双指针 left和right向中间逼近 同时实时维护一个max_water 移动指针的标准是短的边移动

### 代码

```python
class Solution:
    def maxArea(self, height):
        left = 0
        right = len(height) - 1
        max_water = 0
        while left < right:
            curr_water = min(height[left], height[right]) * (right - left)
            max_water = max(max_water, curr_water)
            if height[left] <= height[right]:
                left += 1
            else:
                right -= 1
        return max_water

height = input()

height = [int(x.strip()) for x in height.split(',')]
solution = Solution()
print(solution.maxArea(height))
```

### 类似问题（881. 船接人）

### 思路

本题必须先排序数组 之后每次循环船都必定+1 同时检测是否两边都能坐人

### 代码

```python
class Solution:
    def numRescueBoats(self, people, limit):
        people.sort()
        left = 0
        right = len(people) - 1
        boat = 0
        while left <= right:
            if people[left] + people[right] <= limit:
                left += 1
            right -= 1
            boat += 1
        return boat

people = input("input an array")
limit = int(input("input an integer"))

people = [int(x.strip()) for x in people.split(',')]
solution = Solution()
print(solution.numRescueBoats(people, limit))
```
