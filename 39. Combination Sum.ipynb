# 组合和

### 思路

1. 使用 stack 模拟回溯，每个状态存 path、起始下标 idx、当前和 total。
2. 若 total == target，将 path 加入答案。
3. 若 total > target，说明路径失败，跳过。
4. 否则从 idx 开始遍历 candidates，尝试加入每一个数字。
5. 每次加入数字后把新的 (path, i, total+num) 压入 stack，继续搜索。

### 代码

```python
# LeetCode 39: Combination Sum (迭代版无递归)

class Solution:
    def combinationSum(self, candidates, target):
        candidates.sort()
        res = []
        
        # stack 保存 (当前组合 path, 当前 index idx, 当前和 total)
        stack = [([], 0, 0)]
        
        while stack:
            path, idx, total = stack.pop()
            
            # 达成目标
            if total == target:
                res.append(path)
                continue
            
            # 超过目标，直接剪枝
            if total > target:
                continue
            
            # 继续扩展
            for i in range(idx, len(candidates)):
                num = candidates[i]
                if total + num > target:
                    break
                stack.append((path + [num], i, total + num))
        
        return res


# 测试运行（Notebook 可直接看到输出）
sol = Solution()
print(sol.combinationSum([2,3,6,7], 7))   # [[2,2,3], [7]]
print(sol.combinationSum([2,3,5], 8))     # [[2,3,3], [3,5]]
print(sol.combinationSum([1], 2))         # [[1,1]]
```

### 类似题目（40. 组合和2）

### 思路

1. 先排序，让相同数字相邻，方便去重。
2. 使用迭代栈模拟回溯，每个状态是 (path, idx, total)。
3. 若 total == target，把 path 加入结果。
4. 遍历从 idx 开始的所有候选；若同层数字重复则跳过。
5. 每次加入 candidates[i] 后推进到下一层 idx = i+1，保证每个数只用一次。

### 代码

```python
class Solution:
    def combinationSum2(self, candidates, target):
        candidates.sort()
        res = []
        
        # stack: (path, next_index, total)
        stack = [([], 0, 0)]
        
        while stack:
            path, idx, total = stack.pop()
            
            if total == target:
                res.append(path)
                continue
            
            if total > target:
                continue

            for i in range(idx, len(candidates)):
                num = candidates[i]

                # 同层去重：i > idx 才能比较
                if i > idx and candidates[i] == candidates[i-1]:
                    continue

                if total + num > target:
                    break
                
                # 因为每个数只能用一次，下层 idx = i+1
                stack.append((path + [num], i + 1, total + num))
        
        return res

sol = Solution()
print(sol.combinationSum2([10,1,2,7,6,1,5], 8))
print(sol.combinationSum2([2,5,2,1,2], 5))
```
