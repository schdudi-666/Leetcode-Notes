```python
class Solution:
    def wordBreak(self, s: str, wordDict: list[str]) -> bool:
        # 1ï¸âƒ£ dp[i] è¡¨ç¤º s[:i] èƒ½å¦è¢«æ‹†åˆ†æˆå­—å…¸ä¸­çš„å•è¯
        dp = [False] * (len(s) + 1)
        dp[0] = True  # ç©ºå­—ç¬¦ä¸²å¯ä»¥è¢«æˆåŠŸæ‹†åˆ†

        # 2ï¸âƒ£ çŠ¶æ€è½¬ç§»
        for i in range(1, len(s) + 1):
            for j in range(i):
                if dp[j] and s[j:i] in wordDict:
                    dp[i] = True
                    break

        # æ‰“å° dp æ•°ç»„ï¼Œå¸®åŠ©ç†è§£æ¯ä¸ªä½ç½®æ˜¯å¦èƒ½è¢«æˆåŠŸæ‹†åˆ†
        print(f"DP çŠ¶æ€å˜åŒ–: {dp}")
        return dp[len(s)]

# ğŸ§ª ç¤ºä¾‹è¿è¡Œ
s = "leetcode"
wordDict = ["leet", "code"]
sol = Solution()
result = sol.wordBreak(s, wordDict)

print(f"\nè¾“å…¥: s = {s}, wordDict = {wordDict}")
print(f"è¾“å‡ºç»“æœ: {result}")
```

ç”¨åŠ¨æ€è§„åˆ’ï¼Œdp[i] è¡¨ç¤ºä»¥ nums[i] ç»“å°¾çš„æœ€é•¿é€’å¢å­åºåˆ—é•¿åº¦ï¼›å¯¹æ¯ä¸ªä½ç½® iï¼Œå‘å‰æŸ¥æ‰¾æ‰€æœ‰ j < i ä¸” nums[j] < nums[i] çš„ä½ç½®ï¼Œå–å…¶ä¸­ dp[j] + 1 çš„æœ€å¤§å€¼ä½œä¸º dp[i]ã€‚æœ€ç»ˆç»“æœæ˜¯ max(dp)

### ä»£ç 

```python
class Solution:
    def lengthOfLIS(self, nums: list[int]) -> int:
        n = len(nums)
        dp = [0] * n

        # 1ï¸âƒ£ åˆå§‹åŒ–ï¼šæ¯ä¸ªå…ƒç´ è‡³å°‘èƒ½æ„æˆé•¿åº¦ä¸º 1 çš„å­åºåˆ—
        for i in range(n):
            dp[i] = 1
            # 2ï¸âƒ£ å¯¹äºæ¯ä¸ª iï¼Œå‘å‰æ‰¾æ‰€æœ‰å¯èƒ½çš„ j
            for j in range(i):
                if nums[j] < nums[i]:
                    dp[i] = max(dp[i], dp[j] + 1)
        
        # 3ï¸âƒ£ è¾“å‡ºç»“æœ
        print(f"DP æ•°ç»„çŠ¶æ€: {dp}")
        return max(dp)

# ğŸ§ª ç¤ºä¾‹è¿è¡Œ
nums = [10, 9, 2, 5, 3, 7, 101, 18]
sol = Solution()
result = sol.lengthOfLIS(nums)

print(f"\nè¾“å…¥: {nums}")
print(f"è¾“å‡ºç»“æœ: {result}")
```
