# 戳气球

### 思路

区间 DP：在开区间 (left, right) 中枚举最后一个被戳的气球 k，
状态转移 dp[left][right] = max(dp[left][k] + dp[k][right] + nums[left]*nums[k]*nums[right])。

### 代码

```python
def maxCoins(nums):
    # 在两端加上1，方便处理边界
    nums = [1] + nums + [1]
    n = len(nums)
    
    # 初始化 DP 数组
    dp = [[0] * n for _ in range(n)]
    
    # 从小区间开始递推
    for length in range(2, n):  # 区间长度至少为2（空区间），最大 n-1
        for left in range(0, n - length):
            right = left + length
            # 枚举最后一个被戳的气球
            for k in range(left + 1, right):
                dp[left][right] = max(
                    dp[left][right],
                    dp[left][k] + dp[k][right] + nums[left] * nums[k] * nums[right]
                )
    
    # dp[0][n-1] 表示开区间 (0, n-1) 的最优解
    print("DP Table:")
    for row in dp:
        print(row)
    
    print("\nMaximum coins:", dp[0][n - 1])
    return dp[0][n - 1]

# 示例运行
maxCoins([3, 1, 5, 8])
```

### 类似题目（516. 最长回文子序列）

### 思路

区间 DP：dp[i][j] 表示 s[i...j] 的最长回文子序列长度。
若两端相等则 +2，否则取左右子区间的最大值。

### 代码

```python
def longestPalindromeSubseq(s: str) -> int:
    n = len(s)
    dp = [[0] * n for _ in range(n)]

    # 初始化：单个字符长度为 1
    for i in range(n):
        dp[i][i] = 1

    # 从后往前枚举 i，保证子问题已计算
    for i in range(n - 1, -1, -1):
        for j in range(i + 1, n):
            if s[i] == s[j]:
                dp[i][j] = dp[i + 1][j - 1] + 2
            else:
                dp[i][j] = max(dp[i + 1][j], dp[i][j - 1])

    # 打印 DP 表
    print("DP Table:")
    for row in dp:
        print(row)

    print(f"\nLongest Palindromic Subsequence length: {dp[0][n - 1]}")
    return dp[0][n - 1]

# 示例运行
longestPalindromeSubseq("bbbab")
```
