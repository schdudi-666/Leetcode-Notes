# 字母异位词分组

### 思路

1. 建立一个字典，用来存储 key（排序后字符串） → 同组异位词列表。
2. 对每个字符串，将其转换为字符数组并排序，得到可以标识异位词类别的 key。
3. 将原字符串 append 到字典对应 key 的列表中。
4. 所有字符串处理完后，字典的每个 value 就是一组异位词。
5. 返回字典所有 value 组成的列表即可完成分组。

### 代码

```python
from collections import defaultdict

def groupAnagrams(strs):
    d = defaultdict(list)
    for s in strs:
        key = ''.join(sorted(list(s)))
        d[key].append(s)
    return list(d.values())

# 测试
strs = ["eat","tea","tan","ate","nat","bat"]
groupAnagrams(strs)
```

### 类似题目（242. 合格异位词）

### 思路

本题只需按上题部分思路即可解出

### 代码

```python
def isAnagram(s, t):
    left = ''.join(sorted(list(s)))
    right = ''.join(sorted(list(t)))
    return left == right

s = "anagram"
t = "nagaram"
isAnagram(s, t)
```
