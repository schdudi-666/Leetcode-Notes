# 根据身高排队

### 思路

每个人是 [h, k]：前面要有 k 个身高 ≥ h 的人。

贪心：让高个子先站队，避免后续被矮个干扰。

排序规则：先按身高 h 降序；同高按 k 升序。

按排序后的顺序，把每个人插入结果数组的第 k 个位置。

因为插入时前面只会有更高或同高的人，正好满足 k 条件。

### 代码

```python
# LeetCode 406. Queue Reconstruction by Height

people = [[7,0],[4,4],[7,1],[5,0],[6,1],[5,2]]

# 按身高降序，k 升序排序
people_sorted = sorted(people, key=lambda x: (-x[0], x[1]))
print("排序后的结果:")
print(people_sorted)

res = []
for p in people_sorted:
    # 按照 k 插入到指定位置
    res.insert(p[1], p)

print("\n最终队列:")
print(res)
```

### 类似问题（354. 俄罗斯套娃）

### 思路

每个信封是二维比较：必须 w 更大且 h 更大才能套。

核心技巧：按 w 升序排；若 w 相同，则按 h 降序排。

原因：防止同宽信封被误算成可套。

排序完后，对高度 h 求最长严格递增子序列（LIS）。

LIS 的长度就是最大套娃数。

### 代码

```python
# LeetCode 354. Russian Doll Envelopes

envelopes = [[5,4],[6,4],[6,7],[2,3]]

# Step 1: 自定义排序：w 升序，如果 w 相同，则 h 降序
env_sorted = sorted(envelopes, key=lambda x: (x[0], -x[1]))
print("排序后：")
print(env_sorted)

# Step 2: 提取高度序列
heights = [h for _, h in env_sorted]
print("\n高度序列：", heights)

# Step 3: 对高度求 LIS
import bisect
lis = []

for h in heights:
    pos = bisect.bisect_left(lis, h)
    if pos == len(lis):
        lis.append(h)
    else:
        lis[pos] = h

print("\nLIS 过程结束后列表：", lis)
print("最大套娃数量 =", len(lis))
```
