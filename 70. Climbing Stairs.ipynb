# 爬楼梯

### 思路

本题的思路是动态规划 使用更简单的方式解决递归空间时间占用大的问题 本题在数学上叫做斐波那契数列

### 代码

```python
def climbStairs(n):
    if n <= 2:
        return n
    a, b = 1, 2
    for i in range(3, n + 1):
        a, b = b, b + a
    return b

print(climbStairs(6))
print(climbStairs(7))
print(climbStairs(8))
```

### 类似题目（746. 最小爬楼梯成本）

### 思路

本题的思路是利用动态规划 一步一步递推出最少花费

### 代码

```python
def minCostClimbingStairs(cost):
    high = len(cost)
    dp = [0] * high
    dp[0], dp[1] = cost[0], cost[1]
    for i in range(2, high):
        dp[i] = min(dp[i-1] + cost[i], dp[i-2] + cost[i])
    return min(dp[-1], dp[-2])

cost = [1, 100, 1, 1, 1, 100, 1, 1, 100, 1]
print(minCostClimbingStairs(cost))
```
