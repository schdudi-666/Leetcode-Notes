# 独特路径

### 思路

用动态规划：设 dp[i][j] 为到达 (i,j) 的路径数，只能从上或左走，所以 dp[i][j]=dp[i-1][j]+dp[i][j-1]，第一行第一列初始化为 1，最终答案是 dp[m-1][n-1]。

### 代码

```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        dp = [[0 for _ in range(n)] for _ in range(m)]

        for i in range(m):
            dp[i][0] = 1
        for j in range(n):
            dp[0][j] = 1

        for i in range(1, m):
            for j in range(1, n):
                dp[i][j] = dp[i-1][j] + dp[i][j-1]

        return dp[m-1][n-1]


solution = Solution()
m, n = 4, 100
result = solution.uniquePaths(m, n)
print(f"Grid {m}x{n} 的不同路径数为：{result}")
```

### 类似题目（63. 独特路径2）

### 思路

用动态规划：
设 dp[i][j] 为到达格子 (i,j) 的路径数。
若该格子有障碍（1），则路径数为 0；否则 dp[i][j] = dp[i-1][j] + dp[i][j-1]。
第一行和第一列单独初始化，一旦遇障碍后面的格子不可达（为 0）。
最终返回右下角 dp[m-1][n-1]。

### 代码

```python
from typing import List

class Solution:
    def uniquePathsWithObstacles(self, obstacleGrid: List[List[int]]) -> int:
        m, n = len(obstacleGrid), len(obstacleGrid[0])

        if obstacleGrid[0][0] == 1:
            return 0

        dp = [[0 for _ in range(n)] for _ in range(m)]

        for i in range(m):
            if obstacleGrid[i][0] == 0:
                dp[i][0] = 1
            else:
                break
        for j in range(n):
            if obstacleGrid[0][j] == 0:
                dp[0][j] = 1
            else:
                break
        for i in range(1, m):
            for j in range(1, n):
                if obstacleGrid[i][j] == 0:
                    dp[i][j] = dp[i-1][j] + dp[i][j-1]
                else:
                    dp[i][j] = 0

        return dp[m-1][n-1]

grid = [
    [0,0,0],
    [0,1,0],
    [0,0,0]
]
solution = Solution()
result = solution.uniquePathsWithObstacles(grid)
print(f"不同路径数为：{result}")
```
