```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        dummy = ListNode(0)
        curr = dummy
        carry = 0
        
        while l1 or l2 or carry:
            x = l1.val if l1 else 0
            y = l2.val if l2 else 0
            carry, val = divmod(x + y + carry, 10)
            curr.next = ListNode(val)
            curr = curr.next
            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next
                
        return dummy.next


# üß© ÊµãËØï‰ª£Á†Å
def build_linked_list(nums):
    dummy = ListNode(0)
    cur = dummy
    for n in nums:
        cur.next = ListNode(n)
        cur = cur.next
    return dummy.next

def print_linked_list(node):
    vals = []
    while node:
        vals.append(str(node.val))
        node = node.next
    print(" -> ".join(vals))

l1 = build_linked_list([2, 4, 3, 5, 8, 6])
l2 = build_linked_list([5, 6, 4, 9, 9, 9])
res = Solution().addTwoNumbers(l1, l2)
print_linked_list(res)
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1, l2):
        stack1, stack2 = [], []

        # Êää‰∏§‰∏™ÈìæË°®ÁöÑÂÄºÂéãÊ†à
        while l1:
            stack1.append(l1.val)
            l1 = l1.next
        while l2:
            stack2.append(l2.val)
            l2 = l2.next

        head = None
        carry = 0

        # ‰ªéÊ†àÈ°∂ÂºÄÂßãÈÄê‰ΩçÁõ∏Âä†
        while stack1 or stack2 or carry:
            x = stack1.pop() if stack1 else 0
            y = stack2.pop() if stack2 else 0
            carry, val = divmod(x + y + carry, 10)
            node = ListNode(val)
            node.next = head
            head = node

        return head


# üß© ÊµãËØïÂáΩÊï∞
def build_linked_list(nums):
    dummy = ListNode(0)
    cur = dummy
    for n in nums:
        cur.next = ListNode(n)
        cur = cur.next
    return dummy.next

def print_linked_list(node):
    vals = []
    while node:
        vals.append(str(node.val))
        node = node.next
    print(" -> ".join(vals))


# Á§∫‰æãËøêË°å
l1 = build_linked_list([7, 2, 4, 3])
l2 = build_linked_list([5, 6, 4])
res = Solution().addTwoNumbers(l1, l2)
print_linked_list(res)
```
