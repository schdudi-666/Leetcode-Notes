# 最长有效括号

### 思路

1. 栈初始放入 -1，表示有效子串的起始边界之前的位置。
2. 遍历字符串，遇到 '(' 就把下标压栈。
3. 遇到 ')' 就弹栈，如果栈空说明匹配失败，把当前下标压入栈作为新的边界。
4. 如果栈不空，说明匹配成功，当前有效长度为 `i - 栈顶下标`。
5. 在整个遍历过程中不断更新最大合法长度 longest。

### 代码

```python
def longest_valid_parentheses(s: str) -> int:
    stack = [-1]
    longest = 0
    
    for i, ch in enumerate(s):
        if ch == '(':
            stack.append(i)
        else:
            stack.pop()
            if stack:
                longest = max(longest, i - stack[-1])
            else:
                stack.append(i)
    
    return longest


# -------------------------
# 测试示例（Notebook 输出）
# -------------------------
tests = [
    "(()",
    ")()())",
    "()(()",
    "()(())",
    "(()())",
    "",
    "(",
    ")",
]

for t in tests:
    print(f"Input: {t!r} → Output: {longest_valid_parentheses(t)}")
```

### 类似问题（921. 最小添加括号）

### 思路

1. 使用 `balance` 记录当前未匹配的左括号数量，`add` 记录需要补的括号数。
2. 遇到 '('，balance++；遇到 ')'，balance--。
3. 如果 balance 变成负数，说明有多余的右括号，需要补一个 '('：add++，并把 balance 调回 0。
4. 遍历结束后，balance>0 表示还有很多未匹配的 '('，需要补 balance 个 ')'。
5. 答案就是 add + balance。

### 代码

```python
def min_add_to_make_valid(s: str) -> int:
    balance = 0   # 未匹配的左括号数量
    add = 0       # 需要补的括号数量

    for ch in s:
        if ch == '(':
            balance += 1
        else:
            balance -= 1

        # 如果 balance < 0，说明右括号太多，需要补一个左括号
        if balance < 0:
            add += 1
            balance += 1  # 相当于补了一个 '('

    return add + balance


# -------------------------------
# Notebook 测试输出
# -------------------------------
tests = [
    ")(",
    "(((",
    "())(()",
    "()",
    "(()())",
    "",
    "()))",
    "((())",
]

for t in tests:
    print(f"Input: {t!r} → Output: {min_add_to_make_valid(t)}")
```
