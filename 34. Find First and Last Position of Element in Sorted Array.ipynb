# 找到排序数组中元素第一次和最后一次出现的位置

### 思路

用二分搜索找到任意一个 target 位置 mid。

如果找不到 mid，直接返回 [-1, -1]。

从 mid 向左扩散，找到 target 第一次出现的位置 l。

从 mid 向右扩散，找到 target 最后一次出现的位置 r。

返回 [l, r] 作为 target 出现的区间。

### 代码

```python
# LeetCode 34: Find First and Last Position of Element in Sorted Array

class Solution:
    def searchRange(self, nums, target):
        if len(nums) < 1:
            return [-1, -1]

        left = 0
        right = len(nums) - 1
        m = -1

        # 第一次二分：找到任意一个 target 的位置
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                m = mid
                break
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1

        # 没找到 target
        if m == -1:
            return [-1, -1]

        # 从 m 向左扩散找第一次出现
        l = m
        while l > 0 and nums[l - 1] == target:
            l -= 1

        # 从 m 向右扩散找最后一次出现
        r = m
        while r < len(nums) - 1 and nums[r + 1] == target:
            r += 1

        return [l, r]


# 测试
sol = Solution()
print(sol.searchRange([5,7,7,8,8,10], 8))   # 输出示例
print(sol.searchRange([1], 1))
print(sol.searchRange([2,2,2,2], 2))
print(sol.searchRange([1,2,3], 4))
```

### 类似题目（35. 找到插入位置）

### 思路

利用二分查找的性质 可以很简单解决

### 代码

```python
# LeetCode 35: Search Insert Position

class Solution:
    def searchInsert(self, nums, target):
        left, right = 0, len(nums) - 1

        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] > target:
                right = mid - 1
            else:
                left = mid + 1
        
        # 二分结束后，left 就是插入点
        return left


# 测试
sol = Solution()

print(sol.searchInsert([1,3,5,6], 5))  # 2
print(sol.searchInsert([1,3,5,6], 2))  # 1
print(sol.searchInsert([1,3,5,6], 7))  # 4
print(sol.searchInsert([1,3,5,6], 0))  # 0
print(sol.searchInsert([2,4,6,8], 9))  # 4
```
