# 路径和3

### 思路

任意节点都可以作为路径起点，因此需要遍历整棵树。

外层 DFS：让每个节点依次成为路径起点。

内层 DFS：从起点向下累加路径和，遇到等于 targetSum 就计数。

每条路径必须连续，但不要求从根开始。

总结果为所有起点的路径计数之和。

### 代码

```python
# LeetCode 437. Path Sum III

# 构造一个测试用的二叉树：
#       10
#      /  \
#     5   -3
#    / \    \
#   3   2    11
#  / \   \
# 3  -2   1

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

root = TreeNode(10,
        TreeNode(5,
            TreeNode(3, TreeNode(3), TreeNode(-2)),
            TreeNode(2, None, TreeNode(1))
        ),
        TreeNode(-3, None, TreeNode(11))
)

targetSum = 8


# ------------------------------
#   两层 DFS 解法（你的写法）
# ------------------------------
def pathSum(root, targetSum):
    cnt = 0

    def in_dfs(node, curr):
        nonlocal cnt
        if not node:
            return
        if curr + node.val == targetSum:
            cnt += 1
        in_dfs(node.left, curr + node.val)
        in_dfs(node.right, curr + node.val)

    def out_dfs(node):
        if not node:
            return
        in_dfs(node, 0)
        out_dfs(node.left)
        out_dfs(node.right)

    out_dfs(root)
    return cnt


# 输出结果
print("目标和 =", targetSum)
print("符合的路径条数 =", pathSum(root, targetSum))
```

### 类似题目（112. 路径和）

### 思路

路径必须从根开始，到叶子结束。

DFS 每深入一层，就把 targetSum 减去节点值。

到叶子时检查是否 targetSum == 节点值。

左右子树只要有一个满足即可返回 True。

没找到时返回 False。

### 代码

```python
# LeetCode 112. Path Sum

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# 构建一个测试树：
#       5
#      / \
#     4   8
#    /   / \
#   11  13  4
#  /  \      \
# 7    2      1

root = TreeNode(5,
        TreeNode(4,
            TreeNode(11,
                TreeNode(7),
                TreeNode(2)
            )
        ),
        TreeNode(8,
            TreeNode(13),
            TreeNode(4, None, TreeNode(1))
        )
)

targetSum = 22


# --------------------------
#     DFS（单层）
# --------------------------
def hasPathSum(root, targetSum):
    if not root:
        return False

    # 如果到叶子节点
    if not root.left and not root.right:
        return root.val == targetSum

    # 往左右子树继续 DFS
    return (hasPathSum(root.left, targetSum - root.val) or
            hasPathSum(root.right, targetSum - root.val))


# 输出结果
print("targetSum =", targetSum)
print("是否存在路径 =", hasPathSum(root, targetSum))
```
