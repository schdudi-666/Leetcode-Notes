# 单独数

### 思路

本题的思路可以使用集合增减完成 也可以通过异或^=解法更优

### 代码

```python
# 单个数字例子
nums = [4, 1, 2, 1, 2]

print("输入数组：", nums)

# -------------------------
# 方法 1：你的集合法（最容易理解）
# -------------------------
def singleNumber_set(nums):
    s = set()
    for x in nums:
        if x in s:
            s.remove(x)
        else:
            s.add(x)
    return s.pop()

# -------------------------
# 方法 2：异或法（最优 O(1) 空间）
# -------------------------
def singleNumber_xor(nums):
    ans = 0
    for x in nums:
        ans ^= x
    return ans

# 输出结果
print("用集合方法找到的 single number:", singleNumber_set(nums))
print("用 XOR 方法找到的 single number:", singleNumber_xor(nums))
```

### 类似题目（137. 单独数2）

### 思路

每个整数的二进制有 32 位，题目要求找出只出现一次的数。

对于每一位（0~31），统计所有数字在该位上 1 的出现次数。

出现三次的数字在某一位的 1 次数一定是 3 的倍数。

对每一位的计数取 mod 3，剩下的 1 来自于那个 只出现一次 的数字。

将每一位的 mod 结果拼成最终整数，这就是出现一次的那个数。

### 代码

```python
# 示例输入
nums = [2, 2, 3, 2]

print("输入数组:", nums)

# ----------------------------
# 1) 打印数组的二进制形式
# ----------------------------
print("\n每个数字的二进制（按8位展示）：")
for n in nums:
    print(f"{n:>3} → {format(n, '08b')}")


# ----------------------------
# 2) 按每一位统计出现次数
# ----------------------------
bit_count = [0] * 32
for n in nums:
    for i in range(32):
        bit_count[i] += (n >> i) & 1

# 展示统计结果（低位→高位）
print("\n每一位的 1 出现次数（bit 0 是最低位）：")
print(bit_count)


# ----------------------------
# 3) 每一位对 3 取模
# ----------------------------
bit_mod = [c % 3 for c in bit_count]
print("\n每一位对3取模后的结果：")
print(bit_mod)


# ----------------------------
# 4) 重建结果数字
# ----------------------------
result = 0
for i in range(32):
    if bit_mod[i] == 1:
        result |= (1 << i)

print("\n最终还原结果：", result)
```
