# 合并区间

### 思路

先按每个区间的起点排序（因为合并必须按顺序处理）。

建立一个结果列表 res，把第一个区间先放进去。

遍历后面的每个区间：

如果 当前区间的 start ≤ res 最后一个区间的 end → 有重叠
→ 更新 res 最后一个区间的 end = max(现有 end, 当前 end)。

如果没有重叠 → 直接把当前区间 push 进 res。

遍历完后，res 就是合并后的区间列表。

### 代码

```python
def merge(intervals):
    intervals.sort(key = lambda x: x[0])
    res = [intervals[0]]
    for i in range(1, len(intervals)):
        if intervals[i][0] <= res[-1][1]:
            res[-1][1] = max(res[-1][1], intervals[i][1])
        else:
            res.append(intervals[i])
    return res
intervals = [[1,3],[2,6],[8,10],[15,18]]
print(merge(intervals))
```

### 类似问题（57. 插入区间）

### 思路

本题的思路直接照搬上题即可

### 代码

```python
def insert(intervals, newInterval):
    intervals.append(newInterval)
    intervals.sort(key = lambda x:x[0])
    res = [intervals[0]]
    for i in range(1, len(intervals)):
        if intervals[i][0] <= res[-1][1]:
            res[-1][1] = max(res[-1][1], intervals[i][1])
        else:
            res.append(intervals[i])
    return res
intervals = [[1,3],[6,9]]
newInterval = [2,5]
print(insert(intervals, newInterval))
```
